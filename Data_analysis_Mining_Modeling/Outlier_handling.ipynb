# 上下边界

```python
# 上边界, 上四分位数
upper_q = df['D'].quantile(0.75)
```

```python
# 下边界, 下四分位数
lower_q = df['D'].quantile(0.25)
```

```python
# 四分位间距
q_int = upper_q - lower_q
```

```python
k = 1.5
```

```python
df[df['D']>lower_q-k*q_int][df['D']<upper_q+k*q_int]
```
