```python
import numpy as np
import pandas as pd
from pandas import Series
import matplotlib.pyplot as plt
```

```python
s = Series([1,2,3,4,5])
```

```python
s
```

```python
s.cumsum()
```

```python
s1 = Series(np.random.randn(1000)).cumsum()
s2 = Series(np.random.randn(1000)).cumsum()
```

```python
s1
```

```python
# s1.plot(kind='bar')
s1.plot(kind='line', grid=True, label='S1', title='This is Series')
s2.plot(label='S2')
plt.legend()
```
