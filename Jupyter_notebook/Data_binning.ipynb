```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

对一个array进行分箱

```python
score_list = np.random.randint(25,100,size=20)
```

```python
score_list
```

```python
bins = [0,59,70,80,100]
```

```python
score_cut = pd.cut(score_list, bins)
```

```python
pd.value_counts(score_cut)
```
