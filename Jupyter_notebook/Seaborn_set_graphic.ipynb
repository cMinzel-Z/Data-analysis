```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 更改背景风格

# axes_style and set_style

```python
x = np.linspace(0,14,100)
```

```python
y1 = np.sin(x)
y2 = np.sin(x+2)*1.25
```

```python
def sinplot():
    plt.plot(x,y1)
    plt.plot(x,y2)
```

```python
sinplot()
```

```python
import seaborn as sns
```

```python
sns.set()
```

```python
sinplot()
```

```python
style = ['darkgrid', 'dark', 'white', 'whitegrid', 'ticks']
```

设置风格

```python
sns.set_style(style[0], {'grid.color':'red'})
```

```python
sinplot()
```

```python
sns.axes_style()
```

初始化设置

```python
sns.set()
```

```python
sinplot()
```
